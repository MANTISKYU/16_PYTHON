기본자료형

파이썬에서는 정수, 실수, 복소수 등 숫자를 표현하는 자료형을 모두 숫자 자료형이라고 표현한다.

```python
# int
num1 = 1
num2 = -1

print(type(num1))
print(type(num2))

#float
num3 = 3.14
num4 = 1.0

print(type(num3))
print(type(num4))

#complex
num5 = 3+4j
print(type(num5))
```

### 02-01-02 숫자 자료형의 연산
* 숫자 자료형간에 기본적인 사칙연산과 파이썬에서만 가능한 특수연산을 할 수 있다

```python
num1 = 10
num2 = 3

print(num1 + num2)
print(num1 - num2)
print(num1 * num2)
print(num1 / num2)
```

```python
#서로 다른 자료형의 피연산자를 가질 경우 정수형을 실수형으로 자동 변환
num3 = 10
num4 = 0.1

print(num3 + num4)
print(num3 - num4)
print(num3 * num4)
print(num3 / num4)
```
